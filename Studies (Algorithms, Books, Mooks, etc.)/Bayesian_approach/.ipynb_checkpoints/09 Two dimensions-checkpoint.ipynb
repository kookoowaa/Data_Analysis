## 2.  스윗

- 스윗은 각 가설의 집합(여기에서는 $\alpha$와 $\beta$의 조합으로 정의

````python
class Paintball(thinkbayes.Suite, thinkbayes.Joint):
    
    def __init__(self, alphas, betas, locations):
        self.locations = locations
        pairs = [(alpha, beta) for alpha in alphas for beta in betas]
        thinkbayes.Suite.__init__(self, pairs)
````

- `thinkbayes.Joint` 클래스는 후술
- `location`은 $x$가 발생할 수 있는 모든 장소의 리스트로 Likelihood가 사용 예정

````python
alphas = range(0, 31)
betas = range(1, 51)
locations = range(0, 31)
suite = Paintball(alphas, betas, locations)
````

## 3. 삼각법(Trigonometry)

- 페인트자국이 남을 $x$ location에 대한 우도 함수가 필요
- 상대방이 모든 각도로 사격할 가능성이 있을 때, 경우의 수는 $\alpha$에서 가장 높고, 멀어질수록 낮아짐
- 이를 삼각법으로 표현하면 다음과 같이 표기할 수 있음:
$$x - \alpha = \beta tan \theta$$
- 이를 $\theta$에 대한 식으로 풀면:
$$\theta =  than^{-1}(\frac{x-\alpha}{\beta})$$
- 따라서 특정 location에서 $\theta$를 구하는 게 가능해짐
- $\theta$에 대한 방적식을 미분하면 아래와 같이 표기 가능:
$$\frac{dx}{d\theta} = \frac{\beta}{cos^2\theta}$$

- 이 미분식은 총격속도에 대한 것으로 벽의 주어진 점을 맞출 확률은 총격속도의 역수에 비례
- 만약 총을 쏜 사람과 벽의 위치 좌표를 안다면 총격 속도를 계산 가능


````python
def StrafingSpeed(alpha, beta, x):
    theta = math.atan2(x - alpha, beta)
    speed = beta / math.cos(theta)**2
    return speed
````

- $\alpha$와 $\beta$는 총을 쏜 사람의 좌표고, $x$는 벽에 난 자국의 위치임
- 결과는 $\theta$에 대한 $x$의 미분값을 반환

___
- 벽의 어떤 위치를 맞출 확률을 나타내는 Pmf는 다음과 같이 계산 가능


````python
def MakeLocationPmf(alpha, beta, locations):
    pmf = thinkbayes.Pmf()
    for x in locations:
        prob = 1.0 / StrafingSpeed(alpha, beta, x)
        pmf.Set(x, prob)
        
    pmf.Normalize()
    return pmf        
````

````python
beta10 = MakeLocationPmf(alpha = 10, beta = 10, locations = locations)
beta20 = MakeLocationPmf(alpha = 10, beta = 20, locations = locations)
beta40 = MakeLocationPmf(alpha = 10, beta = 40, locations = locations)
````

````python
sns.lineplot(x = list(beta10.d.keys()), y = list(beta10.d.values()), label = 'beta = 10')
sns.lineplot(x = list(beta20.d.keys()), y = list(beta20.d.values()), label = 'beta = 20')
sns.lineplot(x = list(beta40.d.keys()), y = list(beta40.d.values()), label = 'beta = 40')
plt.title('PMF of locations given alpha = 10\n')
````

## 4. 우도
-`MakeLocationPmf`를 사용해서 $x$ location 값에 대한 우도를 계산
```python
def Likelihood(self, data, hypo):
    alpha, beta = hypo
    x = data
    pmf = MakeLocationPmf(alpha, beta, self.locations)
    like = pmf.Prob(x)
    return like
```

- $\alpha$와 $\beta$는 상대방에 대한 가설(좌표)이고, $x$는 페인트 자국 위치임
- pmf는 ($\alpha$, $\beta$)에 대한 확률이 들어 있음
- 스윗을 갱신할 때는 `UpdateSet`을 사용하면 됨
```python
suite.UpdateSet([15, 16, 18, 21])
```

````python
class Paintball(thinkbayes.Suite, thinkbayes.Joint):
    
    def __init__(self, alphas, betas, locations):
        self.locations = locations
        pairs = [(alpha, beta) for alpha in alphas for beta in betas]
        thinkbayes.Suite.__init__(self, pairs)
        
    def StrafingSpeed(alpha, beta, x):
        theta = math.atan2(x - alpha, beta)
        speed = beta / math.cos(theta)**2
        return speed

    def MakeLocationPmf(alpha, beta, locations):
        pmf = thinkbayes.Pmf()
        for x in locations:
            prob = 1.0 / StrafingSpeed(alpha, beta, x)
            pmf.Set(x, prob)
        
        pmf.Normalize()
        return pmf        
    
    def Likelihood(self, data, hypo):
        alpha, beta = hypo
        x = data
        pmf = MakeLocationPmf(alpha, beta, self.locations)
        like = pmf.Prob(x)
        return like
````

## 5. 결합분포

- 분포의 각 값이 튜플같은 형식으로 되어 있을 때에는 변수가 함께 있는 상태의 분포를 나타내므로 **결합분포<sup>joint distribution</sup>** 이라고 함
- 결합분포를 사용해서 각 변수의 분포를 독립적으로 계산 가능하고 이때의 분포를 **주변분포<sup>marginal distribution</sup>**이라 함
- `thinkbayes.Joint`에서는 주변분포를 다음과 같이 계산
```python
#class Joint:
    def Marginal(self, i):
        pmf = Pmf()
        for vs, prob in self.Items():
            pmf.Incr(vs[i], prob)
        return pmf
    ```

- `i`는 구하고자하는 변수의 인덱스로, `i=0`은 $\alpha$의 분포를, `i=1`은 $\beta$의 분포를 나타냄

````python
suite = Paintball(alphas, betas, locations)
````

````python
suite.UpdateSet([15, 16, 18, 21])
````

````python
marginal_alpha = suite.Marginal(0)
marginal_beta = suite.Marginal(1)
````

````python
sns.lineplot(x = list(marginal_alpha.d.keys()), y = list(marginal_alpha.d.values()), label = 'alpha')
sns.lineplot(x = list(marginal_beta.d.keys()), y = list(marginal_beta.d.values()), label = 'beta')
plt.vlines(x = marginal_alpha.MaximumLikelihood(), ymin = 0, ymax = marginal_alpha.MaxLike(), label = 'most likely alpha', color = 'darkblue')
plt.text(x = marginal_alpha.MaximumLikelihood(),  y = marginal_alpha.MaxLike()*1.01, s = round(marginal_alpha.MaximumLikelihood(),2))
plt.vlines(x = marginal_beta.MaximumLikelihood(), ymin = 0, ymax = marginal_alpha.MaxLike(), label = 'most likely beta', color = 'firebrick')
plt.text(x = marginal_beta.MaximumLikelihood()*1.1,  y = marginal_beta.MaxLike()*1.01, s = round(marginal_beta.MaximumLikelihood(),2))
plt.vlines(x = marginal_beta.Mean(), ymin = 0, ymax = marginal_alpha.MaxLike(), label = 'mean beta', linestyles=':', color = 'firebrick')
plt.text(x = marginal_beta.Mean()*1.01,  y = 0.02, s = round(marginal_beta.Mean(),2))
plt.legend()
plt.title('Posterior PMF of alpha and beta\n')
plt.ylim(0,0.11)
plt.xlim(0,50)
````

- 위 분포를 보면 alpha는 중간값이 약 17로 관측값[15, 16, 18, 21]의 중간 지점임
- beta는 벽 근처에서 가장 우도가 높게 나옴
- 따라서 상대방은 벽 근처에 있을 확률이 높지만, 아래를 보면, 90% 신뢰구간이 방 대부분을 포함하는 것으로 보아 추가 정보가 필요

````python
[marginal_alpha.CredibleInterval(50), marginal_beta.CredibleInterval(50)],
[marginal_alpha.CredibleInterval(90), marginal_beta.CredibleInterval(90)]
````

## 6. 조건 분포

- 주변분포는 변수에 대한 정보를 독립적으로 담고 있지만, 종속성에 대해서는 알 수 없음
- 변수간의 종속성을 시각화 하는 방법 중 하나는 **조건분포<sup>conditional distribution</sup>**를 계산하는것임(`thinkbayes.Joint`)
```python
#thinkbaeys.joint
    def Conditional(self, i, j val):
        pmf = Pmf()
        for vs, prob in self.Items():
            if vs[j]!=val: continue
            pmf.Incr(vs[i], prob)
            
        pmf.Normalize()
        return pmf
        ```

- i는 사용하는 변수의 인덱스고, j는 조건 변수의 인덱스, val은 조건값임
- 위 함수의 결과값은 j번째 변수의 값이 `val`이라는 조건 하에 i번째 변수의 분포임

````python
betas = [10, 20, 40]
for beta in betas:
    cond = suite.Conditional(0, 1, beta)
    sns.lineplot(x = list(cond.d.keys()), y = list(cond.d.values()), label = 'bata =  {}'.format(beta))
````

- **변수가 독립적일 경우 조건 분포는 항상 동일**
- 분포가 다른 경우 변수가 상호 의존적
- 예를 들어 $beta = 10$이라는 걸 알고 있다면, $\alpha$의 조건 분포는 꽤 좁을 것

## 7. 신뢰구간

- 1차원의 경우, 중심으로부터 신뢰구간을 선택하는 것이 일반적임
- 다차원의 경우 특정 신뢰구간이 될수 있을 가능성이 가장 높을 값들을 포함하는 최대 우도 신뢰구간을 선택하는 것이 일반적
- `thinkbayes.Joint`에는 최대 우도 신뢰구간을 계산하는 메서드가 포함
```python
#class Joint:

    def MaxLikeInterval(self, percentage = 90):
        interval = []
        total = 0
    
        t = [(prob, val) for val, prob in self.Items()]
        t.sort(reverse = True)
    
        for prob, val in t:
            intervalappend(val)
            total+=prob
            if total>= percentage/100:
                break
        
        return interval
        ```

- 위 함수는 확률이 큰 것부터 순서대로 나열하여, 전체 확률이 `percentage`를 초과할 때 까지 확률 값을 구간에 더하는 방식을 사용
- 결과는 suite 내 값의 리스트임

````python
def MakeCrediblePlot(suite):
    d = dict((pair, 0) for pair in suite.Values())
    
    percentages = [75, 50, 25]
    for p in percentages:
        interval = suite.MaxLikeInterval(p)
        for pair in interval:
            d[pair] += 1
    
    return d
````

````python
suite_interval = MakeCrediblePlot(suite)
````

````python
sns.scatterplot(x = [i[0] for i in suite_interval.keys()],
                y = [i[1] for i in suite_interval.keys()], 
                hue = list(suite_interval.values()), 
                palette='Blues',
               s = 70,
               marker = 's')
plt.xlim(0,30)
plt.xlabel('alpha')
plt.ylim(0,50)
plt.ylabel('beta')
plt.legend(['75+', '50~75', '25~50', '25-'])

plt.title('Credible interval for the location of shotter\n')
````

- 25% 신뢰구간에서 상대편은 alpha 15~20, beta 0~6 사이에 있을 것으로 추정 가능 
- 신뢰구간이 커질수록 경기장의 오른쪽으로 치우쳐 가는 모습을 볼 수 있음
